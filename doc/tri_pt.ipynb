```
A = Table[a[i, j], {i, 3}, {j, 3}]
T = Table[t[i], {i, 3}]
dA = Table[da[i, j], {i, 3}, {j, 3}]
dT = Table[dt[i], {i, 3}]
```

```
B = Table[b[i, j], {i, 3}, {j, 3}]
U = Table[u[i], {i, 3}]
dB = Table[db[i, j], {i, 3}, {j, 3}]
dU = Table[du[i], {i, 3}]
```

```
pt0 = Table[vp[i], {i, 3}]
ta0 = Table[va[i], {i, 3}]
tb0 = Table[vb[i], {i, 3}]
tc0 = Table[vc[i], {i, 3}]
```

```
Coplanar[x1_, x2_, x3_, x4_] := 0 == Dot[Cross[x2-x1, x4-x1], x3-x1]
```

```
point := (dA . pt0 + dT) * tau + A . pt0 + T
```

```
tri[v_] := (dB . v + dU) * tau + B . v + U
```

```
ta = tri[ta0]
tb = tri[tb0]
tc = tri[tc0]
```

```
ab = tb - ta
bc = tc - tb
ca = ta - tc
ap = point - ta
bp = point - tb
cp = point - tc
```

```
orth = Cross[ab, bc]
```

```
inab = Dot[orth, Cross[ab, ap]]
```

```
inbc = Dot[orth, Cross[bc, bp]]
```

```
inca = Dot[orth, Cross[ca, cp]]
```

```
{solge, solle} = Parallelize[{
    ArgMin[
        {
            tau, 
            tau >= 0,
            Coplanar[point, ta, tb, tc],
            inab >= 0,
            inbc >= 0,
            inca >= 0
        },
        tau,
        Reals
    ],
    ArgMin[
        {
            tau, 
            tau >= 0,
            Coplanar[point, ta, tb, tc],
            inab <= 0,
            inbc <= 0,
            inca <= 0
        },
        tau,
        Reals
    ]
}]
```
